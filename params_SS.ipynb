### Check highest SScore

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
param_N = pd.read_csv("dataset/parameters_N.csv")
param_Y = pd.read_csv("dataset/parameters_Y.csv")
```

```python
dampings = []
max_sss = []
for i, values in param_N.groupby('Damping'):
    damping = i
    max_ss = max(values['Sscore'])
    dampings.append(i)
    max_sss.append(max_ss)
```

```python
dampings
```

```python
max_sss
```

```python
plt.plot(dampings, max_sss)
plt.scatter(dampings, max_sss)
plt.title("Silhouette Score for Each Damping \n (Colonias without Public Water Service)")
plt.xlabel("Damping")
plt.ylabel("Silhouette Score")
max_index = np.argmax(max_sss)
plt.plot(dampings[max_index], max_sss[max_index], 'gs')
text = "Damping=0.6\nNumber of Clusters=11" + "\nSilhouette Score=" + str(round(max_sss[max_index], 2))
print(text)
print("Pos:", dampings[max_index], max_sss[max_index])
plt.annotate(text, xy=(dampings[max_index], max_sss[max_index]), xytext=(0.558, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.0, headwidth=5.0))
plt.savefig("params_N.png")
```
